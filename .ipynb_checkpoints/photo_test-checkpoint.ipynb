## Make or remake sqlite database

```python
import glob

files=[]
root_dir = '/limestone/Data/Photos/web_library/large'
for filename in glob.iglob(root_dir + '**/**', recursive=True):
     files.append(filename)

photos = [f for f in files if ('.jpg') in f[-4:].lower()]
len(photos)   
```

```python
import sqlite3
# os.remove("photo_library.db")

conn = sqlite3.connect('photo_library.db')
c = conn.cursor()
c.execute('''CREATE TABLE photos
             (filename, keywords, comment)''')
```

```python

# Create table
for i in tqdm(range(len(photos))):
    info = IPTCInfo(photos[i])
    keywords = [a.decode("utf-8")  for a in info['keywords']]
    keywords = ' '.join(keywords[0].split(','))
    caption = info['caption/abstract']

    entry = (photos[i],keywords,caption)
    # Insert a row of data
    c.execute("INSERT INTO photos VALUES (?,?,?)", entry)

# Save (commit) the changes
conn.commit()

# We can also close the connection if we are done with it.
# Just be sure any changes have been committed or they will be lost.
# conn.close()
```

```python
# import sqlite3
# os.remove("lib_from_darktable.db")

conn_make = sqlite3.connect('lib_from_darktable.db')
c_make = conn_make.cursor()
c_make.execute('''CREATE TABLE photos
             (filename, keywords, w, h, aspect_ratio, timestamp)''')
```

```python
conn_lib = sqlite3.connect('/home/blake/.config/darktable/library.db')
c_lib = conn_lib.cursor()
conn_data = sqlite3.connect('/home/blake/.config/darktable/data.db')
c_data = conn_data.cursor()

# result = c_data.execute("SELECT * FROM tags WHERE name LIKE ?",('%'+'jpg'+'%',)).fetchall()
result = c_data.execute("SELECT * FROM tags").fetchall()

tag_dict = {}
tag_dict['id']=[]
tag_dict['tag']=[]
for item in result:
    tag_dict['id']+=[item[0]]
    tag_dict['tag']+=[item[1]]

conn_data.close()

result = c_lib.execute("SELECT * FROM tagged_images").fetchall()

tag_img_dict = {}
tag_img_dict['id']=[]
tag_img_dict['tag']=[]
for item in result:
    tag_img_dict['id']+=[item[0]]
    tag_img_dict['tag']+=[item[1]]
    
# result = c_lib.execute("SELECT * FROM images").fetchall()
df = pd.read_sql_query("SELECT * FROM images", conn_lib)

conn_lib.close()

```

```python
#takes 10-15 minutes
list_to_exclude = ['changed','cr2','cr2_default','darktable','default','dng','exif-crop','exported','format','jpg','mode','nef','notes','orf','orf_default','style']

# Create table
for file_id in tqdm(df['id']):

    df_filter = df[df['id']==file_id]

    #make file folder structure
    pre = df_filter['datetime_taken'].values[0].split(' ')[0].replace(':','/')
    post = df_filter['datetime_taken'].values[0].split(' ')[1].replace(':','_')+'.jpg'
    filename = pre+'/'+post
    
    #collect all tags
    if file_id in tag_img_dict['id']:
        tag=[]
        file_tag_ids = [i for i,val in enumerate(tag_img_dict['id']) if val==file_id]
        tags_num = [tag_img_dict['tag'][x] for x in file_tag_ids]
        for k in tags_num:
            loc = tag_dict['id'].index(k)
            tag+=tag_dict['tag'][loc].split('|')
        keywords=' '.join(set(tag) - set(list_to_exclude))
    
    
    #add data to database
    #can add others, caption metadata needs to be grabbed like tags
    entry = (filename,
             keywords,
             df_filter['width'].values[0],
             df_filter['height'].values[0],
             df_filter['aspect_ratio'].values[0],
             df_filter['datetime_taken'].values[0]) 

    c_make.execute("INSERT INTO photos VALUES (?,?,?,?,?,?)", entry)

# Save (commit) the changes
conn_make.commit()

# We can also close the connection if we are done with it.
# Just be sure any changes have been committed or they will be lost.
conn_make.close()
```

```python
from PIL import Image
import io
import matplotlib.pyplot as plt

test='/limestone/Data/Sections/B1123/ortho.tif'
image = Image.open(test)
    
image.thumbnail((image.size[0]/200,image.size[1]/200)) 


buffered = BytesIO()
image.save(buffered, format="TIFF")
img_str = base64.b64encode(buffered.getvalue())

img_base64 = bytes("data:image/tiff;base64,", encoding='utf-8') + img_str



# Take in base64 string and return a numpy image array
def stringToRGB(base64_string):
    imgdata = base64.b64decode(base64_string)
    image = Image.open(io.BytesIO(imgdata))
    return np.array(image)

plt.imshow(stringToRGB(img_base64.decode("utf-8")))
```
